```python
import pandas as pd
import numpy as np
```

```python
import folium

print('Folium installed and imported!')
```

```python
df = pd.read_csv('sda.csv')
```

```python
df.head()
```

```python
# get the first 100 crimes in the df dataframe
limit = 100
df = df.iloc[0:limit, :]
```

```python
df.shape
```

```python
# define the world map
world_map = folium.Map()

# display world map
world_map
```

```python
#define India's geolocation coordinates
ind_latitude = 20.5937
ind_longitude = 78.9629

# define the world map centered around Canada with a higher zoom level
ind_map = folium.Map(location=[ind_latitude, ind_longitude], zoom_start=4)

# display world map
ind_map
```

```python
df['City'].value_counts()
```

```python
geolocator = Nominatim(user_agent="city_locator")
```

```python
cities = df['City'].unique()
```

```python
for city in cities:
    location = geolocator.geocode(f"{city}, India")
    
    if location:
        lat, lon = location.latitude, location.longitude
        # Add marker for each city
        folium.Marker([lat, lon], popup=city).add_to(ind_map)
    else:
        print(f"City not found: {city}")
```

```python

# Filter the dataset for Chennai
chennai_data = df[df['City'] == 'Chennai']

# Get the count of rows where the city is Chennai
chennai_count = chennai_data.shape[0]

# Print the count
print(f"Count of Chennai in the dataset: {chennai_count}")

```
